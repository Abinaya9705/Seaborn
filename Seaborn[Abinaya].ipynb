```python
from sklearn.datasets import fetch_california_housing
```

```python
fetch_housing=fetch_california_housing()
```

```python
house=pd.DataFrame(data=fetch_housing.data,columns=fetch_housing.feature_names)
```

```python
house.head()
```

```python
house.tail()
```

```python
house.shape
```

```python
house.info()
```

```python
house.describe()
```

```python
house.isnull().sum()
```

```python
house.describe(include='all')
```

```python
house.columns
```

```python
house['PRICE']=fetch_housing.target
print(house)
```

```python
sns.distplot(house['PRICE'])
```

```python
correlation=house.corr()
```

```python
correlation
```

```python
sns.heatmap(correlation,annot=True)
```

```python
sns.heatmap(correlation,annot=True,cmap='Blues')
```

```python
sns.heatmap(correlation,annot=True,cmap='Greens')
```

```python
sns.heatmap(correlation,annot=True,cmap='Reds')
```

```python
sns.heatmap(correlation,annot=True,cmap='Purples')
```
